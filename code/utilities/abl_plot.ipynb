```python
# plot with matplotlib, sea
import matplotlib.pyplot as plt
plt.rcParams.update({
    'axes.labelsize': 18,
    'axes.titlesize': 18,
    'legend.fontsize': 18,
    'xtick.labelsize': 18,
    'ytick.labelsize': 18
})
import seaborn as sns
import numpy as np
import sys
import os
minimal = 0.3
highest = 1.1
x_plot = [1,2,3,4]
```

```python
cluster_number = np.array([7,8,9,10])
two_pairs = np.array([0.70, 0.60, 0.7, 0.70])
one_red = np.array([1, 1, 0.81, 0.94])
one_tri = np.array([1, 1, 0.77, 1])

plt.figure(figsize=(5, 5))
# plot points with seaborn
# adjust the marker larger 
# adjust the weight of the lines
# x axis is equal length when plot
sns.lineplot(x=x_plot, y=two_pairs, label='Two Pairs', color='#0072B2', marker='o', markersize=14, linewidth=2, linestyle='--')
sns.lineplot(x=x_plot, y=one_red, label='One Red', color='#E69F00', marker='o', markersize=14, linewidth=2, linestyle='--')
sns.lineplot(x=x_plot, y=one_tri, label='One Triangle', color='#009E73', marker='o', markersize=14, linewidth=2, linestyle='--')
plt.xticks(ticks=x_plot, labels=cluster_number)
plt.yticks([0.5, 0.6, 0.7, 0.8, 0.9, 1.0])  # or generate programmatically
# y base from 0
plt.ylim(minimal, highest)
# plt.title("Comparison of Number of Clusters")
plt.xlabel("Number of Clusters")
plt.ylabel("Accuracy")
plt.grid()
plt.legend(loc='lower right')
# save with the pdf format
plt.savefig("cluster_comparison.pdf", format='pdf', bbox_inches='tight')
# show the plot
plt.show()
plt.close()
```
